```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("airlines_flights_data.csv",)
```

```python
df.info()
df

```

```python
df.drop(columns="index",inplace=True)
```

```python
df
```

```python
df["airline"].nunique()
df["airline"].unique()
```

```python
airline =df["airline"].value_counts()
airline
```

Airline Frequency

```python
y=["Vistara" ,"Air_India" , "Indigo" , "GO_FIRST" ,"AirAsia"  ,"SpiceJet"]
plt.bar(y,airline,color=['r','b','g','m'],label="freq")
plt.legend()
plt.title("Airline Frequency")
plt.ylabel("Number of flights")
plt.xlabel("Airlines")
plt.show()
```

```python
plt.figure(figsize=(16,3))

plt.subplot(1,2,1)
plt.bar(df["departure_time"].value_counts().index,df["departure_time"].value_counts().values,color=['r','g'])
plt.ylabel("number of fight")
plt.xlabel("departure_time")

plt.subplot(1,2,2)
plt.bar(df["arrival_time"].value_counts().index,df["arrival_time"].value_counts().values,color=["y","m"])
plt.ylabel("number of fight")
plt.xlabel("arrival_time")
```
